```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import json

movies = pd.read_csv('/workspaces/cesarseneca-K-nearest/data/raw/tmdb_5000_movies.csv')
credits = pd.read_csv('/workspaces/cesarseneca-K-nearest/data/raw/tmdb_5000_credits.csv')
```

```python
print("Primeras 10 filas de Movies: \n" , movies.head())
print("Forma del dataset" ,movies.shape)
print("Primeras 10 filas de Credits: \n" , credits.head())
print("Forma del dataset", credits.shape)
```

```python
import sqlite3
conn = sqlite3.connect('moviesDB.sqlite')
conn.execute("DROP TABLE IF EXISTS movies_table")
conn.execute("DROP TABLE IF EXISTS credits_table")

movies.to_sql('movies_table', conn, index=False)
credits.to_sql('credits_table', conn, index=False)

query = '''
    SELECT *
    FROM movies_table
    INNER JOIN credits_table
    USING (title);
'''

moviesDB = pd.read_sql_query(query, conn)
conn.close()
```

```python
print("Primeras 10 filas de Movies: \n" , moviesDB.head())
print("Forma del dataset" ,moviesDB.shape)
```

```python
moviesDB = moviesDB[["movie_id","title","overview","genres","keywords","cast","crew"]]
moviesDB
```

```python
moviesDB.isnull().sum()
```

```python
moviesDB.dropna(inplace = True)
```

```python
moviesDB["genres"] = moviesDB["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
moviesDB["keywords"] = moviesDB["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
moviesDB["cast"] = moviesDB["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)
moviesDB["crew"] = moviesDB["crew"].apply(lambda x: next((item["name"] for item in json.loads(x) if item["job"] == "Director"), None) if pd.notna(x) else None)
moviesDB["overview"] = moviesDB["overview"].apply(lambda x: [x] if pd.notna(x) else None)
moviesDB
```

```python
def remove_spaces_from_list(lst):
    return [item.replace(" ", "") for item in lst] if isinstance(lst, list) else lst

moviesDB[["genres", "cast", "keywords"]] = moviesDB[["genres", "cast", "keywords"]].applymap(remove_spaces_from_list)
moviesDB["crew"] = moviesDB["crew"].str.replace(" ", "")

print(moviesDB.head())
```

```python
def list_to_string(lst):
    return ",".join(lst) if isinstance(lst, list) else ""

moviesDB["genres"] = moviesDB["genres"].apply(list_to_string)
moviesDB["keywords"] = moviesDB["keywords"].apply(list_to_string)
moviesDB["cast"] = moviesDB["cast"].apply(list_to_string)
moviesDB["overview"] = moviesDB["overview"].apply(list_to_string)
moviesDB["overview"] = moviesDB["overview"].str.replace(",", "")
```

```python
moviesDB["tags"] = moviesDB["overview"] + " " + moviesDB["genres"] + " " + moviesDB["keywords"]  + " " + moviesDB["cast"] + " " + moviesDB["crew"]
moviesDB["tags"] = moviesDB["tags"].str.replace(",", " ")
```

```python
moviesDB.iloc[0].tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer

moviesDB["tags"] = moviesDB["tags"].fillna("")
cv = CountVectorizer(max_features=5000 ,stop_words='english')
vectors = cv.fit_transform(moviesDB['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    user_words = movie.lower().split()
    mask = moviesDB["title"].str.lower().str.contains('|'.join(user_words))
    matched_movies = moviesDB[mask]

    if matched_movies.empty:
        print("Â¡PelÃ­cula no encontrada!")
        return

    movie_index = matched_movies.index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    print(f"Recomendaciones para pelÃ­culas con las palabras '{movie}':")
    for i, dist in movie_list:
        recommended_movie = moviesDB.iloc[i].title
        print(f"{recommended_movie} (Similitud: {dist:.2f})")

elegir = input("Ingrese el nombre de la pelÃ­cula: ")
recommend(elegir)
```
